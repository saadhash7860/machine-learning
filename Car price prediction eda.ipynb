```python
import seaborn as sns
```

#### Exploratory Data Analysis

```python
count = df["CarName"].value_counts()
print(count)
```

#### Fuel Type

```python
sns.countplot(x= "fueltype",data = df,color= "yellow")
```

#### More than 175 cars uses gas as its fueltype.
#### Very few cars uses diesel fuel.

```python
sns.set(style='whitegrid')
sns.distplot(df['price'],kde = True,color = "red")
plt.show()
```

#### Price distribution cars is right skewed
#### Most of the cars in range of 20000

#### Engine Location

```python
sns.countplot(x = "enginelocation",data = df,color = "salmon")
plt.show()
```

#### Around 200 cars have front engine location and only a few cars have rear location.

```python
count = (df["enginelocation"].value_counts())
print(count)
```

#### Our assumption about car engine location was right which we made from the countplot.

```python
df.corr()
```

#### DoorNumber

```python
count = (df['doornumber'].value_counts())
print(count)
```

```python
plt.figure(figsize = (5,5))
plot = count.plot.pie(y='doornumber')
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.legend()
plt.show()
```

####  Out of 205 cars, 115 cars have four door and 90 cars has two doors.

#### Aspiration

```python
sns.countplot(x = "aspiration",data = df,color = "green")
plt.show()
```

```python
count = (df["aspiration"].value_counts())
print(count)
```

#### 168 cars have std aspiration and only 37 cars have turbo aspiration.

#### Turbo Engines are most expensive as comapred to std(standard engine)
