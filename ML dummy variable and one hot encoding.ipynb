## Using Pandas to create dummy variables.

```python
dummies = pd.get_dummies(df["Car Model"])
dummies
```

```python
dummies
```

```python
Merge = pd.concat([df,dummies],axis = 1)
Merge
```

## Dummy Trap:-

##### When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of Audi A5 and BMW X5 then you can easily infer value of Mercedy Benz C class state, i.e. BMW X5=1 and  =  Audi A5. There for these state variables are called to be multi-colinear. In this situation linear regression won't work as expected. Hence you need to drop one column.

```python
#To prevent dummy trap as we have k = 3 dummy columns we will keep k-1 columns.
final = Merge.drop(["Car Model", "Mercedez Benz C class"],axis = 1)
final
```

###  One-Hot Encoding with Scikit-learn

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#creating a object for class(OneHotEncoder)
oh_en = OneHotEncoder(sparse = False)
```

```python
oh_enc_arr =oh_en.fit_transform(df[["Car Model"]])
oh_enc_arr
```

```python
Merge.keys()
```

```python
oh_enc_arr_df = pd.DataFrame(oh_enc_arr,columns = ['Audi A5','BMW X5', 'Mercedez Benz C class'])        
oh_enc_arr_df
```

```python
dfonehot = pd.concat([df,oh_enc_arr_df],axis = 1)
dfonehot.head()
```
