```python
#  Add a new column using a dictionary.
# Use an existing column as the key values and their respective values will be the values for a new column. 

dict = {"name":["rahul","ali","harry","marie"],"class":[1,3,10,11],"school":["dps","abc","trv","dav"]}
df = pd.DataFrame(dict)
df
dicti = {"marks" :"rahul","kota":"ali","sww":"harry","trrr":"marie"}
df["dicti"] = dicti
df
```

```python
#Adding a new column using list
dict = {"name":["rahul","ali","harry","marie"],"class":[1,3,10,11],"school":["dps","abc","trv","dav"]}
df = pd.DataFrame(dict)
df
address = ["chennai","mumbai","bihar","karnataka"]    #assigning new column name by using list
df["address"] = address
df
```

```python
#Adding a new column using .assign function

dict = {"name":["rahul","ali","harry","marie"],"class":[1,3,10,11],"school":["dps","abc","trv","dav"]}
df = pd.DataFrame(dict)
df
df = df.assign(address = ["chennai","mumbai","bihar","karnataka"])
df
```
