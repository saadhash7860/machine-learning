#### Filling null values using fillna() , replace() , interpolate()

```python
df
```

```python
df2 = df.fillna(value = 0) # where ever the value was Nan ,now its filled with 0.
df2
```

```python
df2.isnull().sum().sum() #We can see now there is no null values in df2 DataFrame.
```

```python
#Filling the null values with previous ones
df3 = df.fillna(method = "pad" )
df3
```

```python
#Filling the null values with next ones.
df4 = df.fillna(method = "bfill")
df4
```

```python
df
```

```python
#Filling the null values with previous ones but column wise.
df5 = df.fillna(method = "pad",axis = 1)
df5
```

```python
#Filling the null values with next ones but column wise.
df6 = df.fillna(method = "bfill",axis =1)
df6
```

#### dropping values using dropna()

```python
df
```

```python
df.dropna() #It simply drops that rows which contain even one null value.
```

```python
df.dropna(how = "any") # it will drop all those rows which even contain one null values.
```

```python
df.dropna(how = "all") # it will drop that row which has all the vaues null.
```

```python
df.dropna(axis=1) #It simply drops that columns which contain even one null value.
```

```python
df.dropna(how = "any",axis = 1) # it will drop all those columns which even contain one null values.
```

```python
df.dropna(how = "all",axis =1)  # it will drop that columns which has all the vaues null.
```

```python
df.dropna(thresh = 1 , axis = 1)
```
